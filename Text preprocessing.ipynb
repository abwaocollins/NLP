```python
#remove punctuations
import string
string.punctuation

def remove_punctuations(text):
  punc_free = ''.join([i for i in text if i not in string.punctuation])
  return punc_free

```

```python
x_data = data['Message']
data['cln_msg'] = x_data.apply(lambda x:remove_punctuations(x))
```

```python
data.head(5)
```

```python
#convert data to lower case
data['lower_txt'] = data['cln_msg'].apply(lambda x:x.lower())
data.head(5)
```

```python
import nltk
from nltk import tokenize
nltk.download('punkt')
```

```python
tokenize.
```

```python
data['tokenized'] = data['lower_txt'].apply(lambda x:tokenize.word_tokenize(x))
data.head(5)

```

```python
#Stop words removal
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
stopwords

def rem_stopwords(text):
  free_stops = [i for i in text if i not in stopwords]
  return free_stops

data['no_stopwords'] = data['tokenized'].apply(lambda x:rem_stopwords(x))
data.head(5)
```

```python
from nltk import stem
lanc = stem.LancasterStemmer()
def stem_text(text):
  output = [lanc.stem(word) for word in text]
  return output

data['stemmed_text'] = data['no_stopwords'].apply(lambda x:stem_text(x))

data.head(5)
```

```python
nltk.download('omw-1.4')
nltk.download('wordnet')
```

```python

word_lem = stem.WordNetLemmatizer()

def lematize(text):
  output = [word_lem.lemmatize(word) for word in text]
  return output

data['lemmatized_text'] = data['no_stopwords'].apply(lambda x:lematize(x))

data.head(5)

```
